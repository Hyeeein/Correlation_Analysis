```python
import os
import sys
import urllib.request

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import json
```

```python
# API 인증정보 설정
client_id = "0QF5tXiOrwFNKLekzOgj"
client_secret = "2hDNhF8Aic"
url = "https://openapi.naver.com/v1/datalab/search";

# body 안의 자세한 내용
# - startDate: 수집 시작일
# - endDate: 수집 종료일
# - timeUnit: 데이터 구간 단위 설정 (date, week, month)
# - keywordGroups: 사용자가 원하는 키워드 입력 (동시에 5개까지 입력 가능)
# - device: PC 검색량, mobile 검색량 선택 가능
# - age: 검색 연령 선택 가능 (단위는 5)
# - gender: 검색 성별

body = "{\"startDate\":\"2019-01-01\",\"endDate\":\"2022-10-31\",\"timeUnit\":\"month\",\
         \"keywordGroups\":[{\"groupName\":\"물티슈\",\"keywords\":[\"물티슈\",\"wet tissue\"]}, \
        {\"groupName\":\"마스크\",\"keywords\":[\"마스크\",\"mask\", \"KF94\"]}, \
        {\"groupName\":\"샴푸\",\"keywords\":[\"샴푸\",\"shampoo\"]}, \
        {\"groupName\":\"오메가3\",\"keywords\":[\"오메가3\", \"omega3\", \"알티지오메가3\"]}], \
         \"device\":\"\", \"ages\":[], \"gender\":\"\"}";

# 네이버 API 서버에 request 요청을 보내고 다시 응답
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
request.add_header("Content-Type","application/json")
response = urllib.request.urlopen(request, data=body.encode("utf-8"))
rescode = response.getcode()

# 응답 받은 데이터가 코드 또는 서버 오류가 발생할 수 있음
# 응답코드를 활용해 if 조건문을 넣음
if(rescode==200):
    response_body = response.read()
    query = response_body.decode('utf-8')
    print(query)
    data_query = json.loads(query)
else:
    print("Error Code:" + rescode)
```

```python
dates = []
wettissue_data = []
for i in data_query['results'][0]['data']:
    dates.append(i['period'])
    wettissue_data.append(i['ratio'])
    
mask_data = []
for i in data_query['results'][1]['data']:
    mask_data.append(i['ratio'])

shampoo_data = []
for i in data_query['results'][2]['data']:
    shampoo_data.append(i['ratio'])

omega3_data = []
for i in data_query['results'][3]['data']:
    omega3_data.append(i['ratio'])

df1 = pd.DataFrame([dates, wettissue_data, mask_data, shampoo_data, omega3_data]).T
df1.columns = ['Date', 'WetTissue', 'Mask', 'Shampoo', 'Omega3']
df1
```
