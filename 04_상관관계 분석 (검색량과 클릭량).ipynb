## 상관계수, 유의수준 검정

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
file_search = pd.read_csv('./Data/네이버 검색어트렌드 8가지 품목 검색량.csv', sep=',')
file_search.head()
```

```python
file_click = pd.read_csv('./Data/네이버 쇼핑인사이트 8가지 품목 클릭량.csv', sep=',')
file_click.head()
```

```python
# 결측치 측정
file_search.isnull().sum()
```

```python
# 결측치 측정
file_click.isnull().sum()
```

```python
# 피어슨 상관계수 검정
# 결과는 순서대로 상관계수, p-value

corr_wettissue = stats.pearsonr(file_search['WetTissue'], file_click['WetTissue'])
print('물티슈:', corr_wettissue)

corr_mask = stats.pearsonr(file_search['Mask'], file_click['Mask'])
print('마스크:', corr_mask)

corr_shampoo = stats.pearsonr(file_search['Shampoo'], file_click['Shampoo'])
print('샴푸:', corr_shampoo)

corr_omega3 = stats.pearsonr(file_search['Omega3'], file_click['Omega3'])
print('오메가3:', corr_omega3)

corr_national = stats.pearsonr(file_search['NationalGeographic'], file_click['NationalGeographic'])
print('내셔널지오그래픽 패딩:', corr_national)

corr_nike = stats.pearsonr(file_search['Nike'], file_click['Nike'])
print('나이키 운동화:', corr_nike)

corr_pokemon = stats.pearsonr(file_search['PokemonBread'], file_click['PokemonBread'])
print('포켓몬빵:', corr_pokemon)

corr_iphone = stats.pearsonr(file_search['iPhone'], file_click['iPhone'])
print('아이폰:', corr_iphone)
```

```python
# 켄달 상관계수 검정
# 결과는 순서대로 상관계수, p-value

corr_wettissue = stats.kendalltau(file_search['WetTissue'], file_click['WetTissue'])
print('물티슈:', corr_wettissue)

corr_mask = stats.kendalltau(file_search['Mask'], file_click['Mask'])
print('마스크:', corr_mask)

corr_shampoo = stats.kendalltau(file_search['Shampoo'], file_click['Shampoo'])
print('샴푸:', corr_shampoo)

corr_omega3 = stats.kendalltau(file_search['Omega3'], file_click['Omega3'])
print('오메가3:', corr_omega3)

corr_national = stats.kendalltau(file_search['NationalGeographic'], file_click['NationalGeographic'])
print('내셔널지오그래픽 패딩:', corr_national)

corr_nike = stats.kendalltau(file_search['Nike'], file_click['Nike'])
print('나이키 운동화:', corr_nike)

corr_pokemon = stats.kendalltau(file_search['PokemonBread'], file_click['PokemonBread'])
print('포켓몬빵:', corr_pokemon)

corr_iphone = stats.kendalltau(file_search['iPhone'], file_click['iPhone'])
print('아이폰:', corr_iphone)
```

```python
# 스피어만 상관계수 검정
# 결과는 순서대로 상관계수, p-value

corr_wettissue = stats.spearmanr(file_search['WetTissue'], file_click['WetTissue'])
print('물티슈:', corr_wettissue)

corr_mask = stats.spearmanr(file_search['Mask'], file_click['Mask'])
print('마스크:', corr_mask)

corr_shampoo = stats.spearmanr(file_search['Shampoo'], file_click['Shampoo'])
print('샴푸:', corr_shampoo)

corr_omega3 = stats.spearmanr(file_search['Omega3'], file_click['Omega3'])
print('오메가3:', corr_omega3)

corr_national = stats.spearmanr(file_search['NationalGeographic'], file_click['NationalGeographic'])
print('내셔널지오그래픽 패딩:', corr_national)

corr_nike = stats.spearmanr(file_search['Nike'], file_click['Nike'])
print('나이키 운동화:', corr_nike)

corr_pokemon = stats.spearmanr(file_search['PokemonBread'], file_click['PokemonBread'])
print('포켓몬빵:', corr_pokemon)

corr_iphone = stats.spearmanr(file_search['iPhone'], file_click['iPhone'])
print('아이폰:', corr_iphone)
```
